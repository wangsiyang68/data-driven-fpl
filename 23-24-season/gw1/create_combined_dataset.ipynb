```python
## import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Train Set

```python
## get defender data from 2021-22 season
## import data from the master folder
data_20_21_train = pd.read_csv(r'C:\Users\Siyang\Documents\projects\fpl\Fantasy-Premier-League\data\2020-21\cleaned_players.csv')
## visualize data
data_20_21_train.head()
```

```python
## get defenders only
selected_data_20_21_train = data_20_21_train[data_20_21_train['element_type'] == 'DEF']
selected_data_20_21_train.index = range(len(selected_data_20_21_train))

## get the form data from the current table only
selected_data_20_21_train = selected_data_20_21_train[['first_name', 'second_name','goals_scored', 'assists', 'total_points',
       'minutes', 'goals_conceded', 'creativity', 'influence', 'threat',
       'bonus', 'bps', 'ict_index', 'clean_sheets', 'red_cards',
       'yellow_cards']]

## Use the full name as a key
selected_data_20_21_train.insert(0, "full_name", selected_data_20_21_train['first_name'] + ' ' + selected_data_20_21_train['second_name'])

## Drop name duplicates
selected_data_20_21_train[selected_data_20_21_train['full_name'].duplicated()]
# selected_data_20_21_train[selected_data_20_21_train.full_name == "Ben Davies"]
selected_data_20_21_train = selected_data_20_21_train.drop(selected_data_20_21_train.index[130])

selected_data_20_21_train.index = range(len(selected_data_20_21_train))
print("Is there any duplicated name in the dataset? ", selected_data_20_21_train['full_name'].duplicated().any())

print("Is there a NaN value in dataframe?", selected_data_20_21_train.isnull().values.any())
print("Number of rows in dataframe:", selected_data_20_21_train.shape[0])
selected_data_20_21_train
```

```python
## get defender data from 2021-22 season
## import data from the master folder
data_21_22_initial_train = pd.read_csv(r'C:\Users\Siyang\Documents\projects\fpl\initial_cleaned_players_21_22.csv')
data_21_22_initial_train
```

```python
## get defenders only
selected_data_21_22_initial_train = data_21_22_initial_train[data_21_22_initial_train['element_type'] == 'DEF']
selected_data_21_22_initial_train.index = range(len(selected_data_21_22_initial_train))

## get the current price and % of ownership for each player
selected_data_21_22_initial_train = selected_data_21_22_initial_train[['first_name', 'second_name', 'now_cost', 'selected_by_percent']]

## combine full name
selected_data_21_22_initial_train.insert(0, 'full_name', selected_data_21_22_initial_train['first_name'] + ' ' + selected_data_21_22_initial_train['second_name'])
selected_data_21_22_initial_train = selected_data_21_22_initial_train.drop(columns=['first_name', 'second_name'], inplace=False)

# ## check for duplicates
selected_data_21_22_initial_train[selected_data_21_22_initial_train['full_name'].duplicated()]
# selected_data_21_22_initial_train[selected_data_21_22_initial_train.full_name == 'Ben Davies']
selected_data_21_22_initial_train = selected_data_21_22_initial_train.drop(selected_data_21_22_initial_train.index[89])
selected_data_21_22_initial_train.index = range(len(selected_data_21_22_initial_train))

print("Is there a NaN value in dataframe?", selected_data_21_22_train.isnull().values.any())
print("Is there any duplicated name in the dataset? ", selected_data_21_22_train['full_name'].duplicated().any())
print("Number of rows in dataframe:", selected_data_21_22_train.shape[0])
selected_data_21_22_initial_train
```

```python
## add final cost at the end of the season
data_21_22_final_train = pd.read_csv(r"C:\Users\Siyang\Documents\projects\fpl\Fantasy-Premier-League\data\2021-22\cleaned_players.csv")

selected_data_21_22_final = data_21_22_final_train[data_21_22_final_train['element_type'] == 'DEF']
selected_data_21_22_final.index = range(len(selected_data_21_22_final))

selected_data_21_22_final = selected_data_21_22_final[['first_name', 'second_name', 'now_cost']]
selected_data_21_22_final.rename(columns={'now_cost': 'final_cost'}, inplace=True)

## insert understat id
selected_data_21_22_final.insert(0, 'full_name', selected_data_21_22_final['first_name'] + ' ' + selected_data_21_22_final['second_name'])
selected_data_21_22_final.drop(columns=['first_name', 'second_name'], inplace=True)

# ## check for duplicates
selected_data_21_22_final[selected_data_21_22_final['full_name'].duplicated()]
selected_data_21_22_final[selected_data_21_22_final.full_name == 'Ben Davies']
selected_data_21_22_final = selected_data_21_22_final.drop(selected_data_21_22_final.index[129])

id_dict = pd.read_csv(r"C:\Users\Siyang\Documents\projects\fpl\Fantasy-Premier-League\data\2021-22\id_dict.csv")
id_dict = id_dict[[' FPL_Name', 'Understat_ID']]
selected_data_21_22_final = selected_data_21_22_final.join(id_dict.set_index(' FPL_Name'), on='full_name')
selected_data_21_22_final = selected_data_21_22_final[~selected_data_21_22_final.Understat_ID.isnull()]
selected_data_21_22_final = selected_data_21_22_final.astype({'Understat_ID': 'int64'})
# selected_data_21_22_final.drop(columns=['full_name'], inplace=True)

## merge initial and final cost
selected_data_21_22_train = selected_data_21_22_initial_train.join(selected_data_21_22_final.set_index('full_name'), on='full_name')
# selected_data_21_22_train.set_index('Understat_ID', inplace=True)

## remove the NaN values, total of 5 found
selected_data_21_22_train = selected_data_21_22_train.dropna()
selected_data_21_22_train.index = range(len(selected_data_21_22_train))

print("Is there a NaN value in dataframe?", selected_data_21_22_train.isnull().values.any())
print("Is there any duplicated name in the dataset? ", selected_data_21_22_train['full_name'].duplicated().any())
print("Number of rows in dataframe:", selected_data_21_22_train.shape[0])
selected_data_21_22_train
```

```python
## combine both selected_data_2122 and selected_data_2223 on the Understad_ID
combined_seasons_data_train = pd.merge(selected_data_20_21_train, selected_data_21_22_train, on='full_name', how='inner')
combined_seasons_data_train = combined_seasons_data_train.astype({'Understat_ID': 'int64'})
combined_seasons_data_train.set_index('Understat_ID', inplace=True)

## drop player names
combined_seasons_data_train.drop(['full_name', 'first_name', 'second_name'], axis=1, inplace=True)

print("Is there a NaN value in dataframe?", combined_seasons_data_train.isnull().values.any())
print("Number of rows in dataframe:", combined_seasons_data_train.shape[0])
print("Number of columns in dataframe:", combined_seasons_data_train.shape[1])
combined_seasons_data_train
```

```python
combined_seasons_data_train.to_csv('combined_seasons_data_train.csv', index=True)
```

```python
## normalize the data
std_train = combined_seasons_data_train.std()
mean_train = combined_seasons_data_train.mean()
norm_combined_seasons_data_train = (combined_seasons_data_train - mean_train) /std_train
norm_combined_seasons_data_train.head()
```
