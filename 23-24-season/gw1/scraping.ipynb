Scraping football statistics from FBREF

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://fbref.com/en/comps/9/2021-2022/stats/2021-2022-Premier-League-Stats"
page = requests.get(url)

# https://stackoverflow.com/questions/72675092/cant-read-html-table-with-pd-read-html
page = page.text.replace('<!--', '')
# soup = BeautifulSoup(page.text, 'html.parser')
stats = pd.read_html(page)
```
